```python
import math

trojkaty = [[1,2],[30,1],[1,3],[5,8],[1,2],[9,10],[9,9]]
wymiary = 0
najwieksza_przeciwprostokatna = 0

for trojkat in trojkaty :
    przeciwprostokatna = math.sqrt(trojkat[0]**2 + trojkat[1]**2)
    if przeciwprostokatna > najwieksza_przeciwprostokatna:
        najwieksza_przeciwprostokatna = przeciwprostokatna
        wymiary = trojkat

print(wymiary)
```
