# Terminal Canvas 🕹️

Rysowanie pikseli w terminalu — jak w starych grach typu Snake.

Żadnych bibliotek graficznych! Tylko tekst i znaki blokowe `█`.

Możesz:
- Zapalać i gasić piksele na siatce
- Rysować linie i prostokąty
- Robić animacje w terminalu

````python
from IPython.display import clear_output
import time
````

## Siatka pikseli

Siatka to zwykła lista list. Każdy element to `0` (zgaszony) albo `1` (zapalony).

Funkcja `render` zamienia siatkę na tekst — `·` to pusty piksel, `█` to zapalony.

````python
def make_grid(width, height):
    return [[0] * width for _ in range(height)]


def render(grid):
    chars = {0: "· ", 1: "██"}
    for row in grid:
        print("".join(chars[pixel] for pixel in row))
````

## Zapalanie pikseli

Żeby zapalić piksel, ustaw `grid[wiersz][kolumna] = 1`.

````python
grid = make_grid(15, 10)

grid[2][5] = 1
grid[4][10] = 1
grid[7][3] = 1

render(grid)
````

## Rysowanie kształtów

Linia pozioma, linia pionowa i prostokąt — wszystko przez ustawianie pikseli w pętli.

````python
grid = make_grid(20, 12)

for x in range(3, 17):
    grid[1][x] = 1

for y in range(2, 10):
    grid[y][0] = 1

for x in range(6, 14):
    grid[4][x] = 1
    grid[9][x] = 1
for y in range(4, 10):
    grid[y][6] = 1
    grid[y][13] = 1

render(grid)
````

## Animacja — poruszający się piksel

`clear_output` czyści ekran przed każdą klatką, a `time.sleep` spowalnia animację.

````python
grid = make_grid(20, 10)

for x in range(20):
    grid = make_grid(20, 10)
    grid[5][x] = 1
    clear_output(wait=True)
    render(grid)
    time.sleep(0.15)
````

## Snake! 🐍

Wąż to lista pozycji. Co klatkę dodajemy nową głowę i usuwamy ogon.

Wąż idzie po prostokątnej ścieżce wokół planszy.

````python
WIDTH, HEIGHT = 20, 12
SNAKE_LEN = 6

path = []
for x in range(WIDTH):
    path.append((0, x))
for y in range(1, HEIGHT):
    path.append((y, WIDTH - 1))
for x in range(WIDTH - 2, -1, -1):
    path.append((HEIGHT - 1, x))
for y in range(HEIGHT - 2, 0, -1):
    path.append((y, 0))

for step in range(len(path)):
    grid = make_grid(WIDTH, HEIGHT)

    for i in range(SNAKE_LEN):
        idx = (step - i) % len(path)
        r, c = path[idx]
        grid[r][c] = 1

    clear_output(wait=True)
    render(grid)
    time.sleep(0.1)
````
