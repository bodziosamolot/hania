```python
#wielkie liczby zaczynaja sie od 65 i koncza sie na 90
def czy_z_wielkiej_litery(slowo):
    pierwsza_litera = slowo[0]
    pierwsza_litera_jako_liczba = ord(pierwsza_litera)
    if pierwsza_litera_jako_liczba > 64 and pierwsza_litera_jako_liczba < 91:
        return True
    else:
        return False
        
wynik = czy_z_wielkiej_litery("KOTEK")
print(wynik)
```

```python
def czy_z_wielkiej_litery_2(slowo):
    pierwsza_litera = slowo[0]
    pierwsza_litera_zamieniona_na_wielka = pierwsza_litera.upper()
    
    return pierwsza_litera == pierwsza_litera_zamieniona_na_wielka

wynik = czy_z_wielkiej_litery_2("Dupa")
print(wynik)
```
