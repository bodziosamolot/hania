```python
def czy_jest_piatka(lista):
    odpowiedz = ''
    for liczba in lista:
        if liczba==5:
            odpowiedz = 'tak'
            break
        else:
            odpowiedz = 'nie'
    
    return odpowiedz
```

```python
def czy_jest_piatka_tata(lista):
    for liczba in lista:
        if liczba==5:
            return "tak"
    
    return "nie"
```

```python
lista = [1,2,3,4,6,7]

czy_jest = czy_jest_piatka(lista)
print(czy_jest)

czy_jest_tata = czy_jest_piatka_tata(lista)
print(czy_jest_tata)
```

```python
def NWW(liczba1, liczba2):
    wielokrotnosci_liczby1 = {}
    wielokrotnosci_liczby2 = {}

    for wielokrotnosc in range(1,100):
        wielokrotnosc_liczby1 = wielokrotnosc * liczba1
        wielokrotnosci_liczby1[wielokrotnosc_liczby1] = True

        wielokrotnosc_liczby2 = wielokrotnosc * liczba2
        wielokrotnosci_liczby2[wielokrotnosc_liczby2] = True

        nww1 = wielokrotnosci_liczby1.get(wielokrotnosc_liczby2)
        if(nww1):
            return wielokrotnosc_liczby2
        
        nww2 = wielokrotnosci_liczby2.get(wielokrotnosc_liczby1)
        if(nww2):
            return wielokrotnosc_liczby1


nww = NWW(7,22)
print(nww)
```
