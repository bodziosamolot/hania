```python
def twoSum(nums, target):
    granica = len(nums)-1
    aktualny_index = 0
    nastepny_index = 1
    
    while granica >= nastepny_index:
        suma = nums[aktualny_index] + nums[nastepny_index]
        if suma == target:
            return [aktualny_index, nastepny_index]
        else:
            nastepny_index=nastepny_index+1
        
        if nastepny_index == granica+1:
            aktualny_index = aktualny_index + 1
            nastepny_index = aktualny_index + 1
            
        
```

```python
wynik = twoSum([1,2,3,4,5,6], 4)
print(wynik)

wynik = twoSum([1,2,3,4,5,6], 8)
print(wynik)
```

```python
wynik = twoSum([3,2,3], 6)
print(wynik)
```
