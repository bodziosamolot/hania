```python
liczby_do_mnozenia = range(1,10)
wiersz = ""

for wiersz_index in liczby_do_mnozenia:
    for liczba in liczby_do_mnozenia:
        mnozonko = wiersz_index * liczba
        wiersz = wiersz + "  " + str(mnozonko)
    print(wiersz)
    wiersz = ""
    
    
```
