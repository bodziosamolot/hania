```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
os_x = []
os_y = []

for x in range(0,2):
    y = x + 2

    os_x.append(x)
    os_y.append(y)

print(os_x)
print(os_y)
```

```python
plt.xlim(0,2)
plt.ylim(0,4)

plt.xticks(np.arange(0, 2, 0.1)) 
plt.yticks(np.arange(0, 4, 0.1)) 

plt.grid(True)
plt.plot(os_x, os_y, label = 'y=x+2')
```

```python
os_x = []
os_y = []

for x in range(0,10):
    y = x*(1/2)+3

    os_x.append(x)
    os_y.append(y)

print(os_x)
print(os_y)
```

```python
plt.xlim(0,10)
plt.ylim(0,10)

plt.xticks(np.arange(0, 10, 1)) 
plt.yticks(np.arange(0, 10, 1)) 

plt.grid(True)
plt.plot(os_x, os_y, label='y=x/2+7')
```
