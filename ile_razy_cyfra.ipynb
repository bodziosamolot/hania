```python
def ile_razy_cyfra(liczba, cyfra):
    ile_razy_w_liczbie = 0
    liczba_przecyferkowana = str(liczba)
    cyfra_jako_slowo = str(cyfra)
    for literka in liczba_przecyferkowana:
        if literka == cyfra_jako_slowo:
            ile_razy_w_liczbie = ile_razy_w_liczbie + 1
        
    return ile_razy_w_liczbie
```
