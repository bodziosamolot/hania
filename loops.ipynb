```python
tygodnie = range(4)
dni = ["poniedzialek","wtorek","sroda","czwartek","piatek","sobota","niedziela"]
godziny = range(0,24)

for tydzien in tygodnie:
    print("tydzien: " + str(tydzien))
    for dzien in dni:
        print("         " + dzien)
        godziny_razem = "            "
        for godzina in godziny:
            godziny_razem = godziny_razem + str(godzina) + "  "
        print(godziny_razem)
```
