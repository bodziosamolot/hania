```python
class Piosenka:
    def __init__(self, tytul):
        self.tytul = tytul
        
class PiosenkaWPlayliscie:
    def __init__(self, piosenka, poprzednia, nastepna):
        self.tytul = piosenka.tytul
        self.nastepna = nastepna
        self.poprzednia = poprzednia
        
class Playlista:
    def __init__(self, nazwa):
        self.nazwa = nazwa
        self.aktywna_piosenka = None
        
    def dodaj(self, piosenka):
        if self.aktywna_piosenka is None:
            self.aktywna_piosenka = PiosenkaWPlayliscie(piosenka, None, None)
        else:
            ostatnia_piosenka = self.aktywna_piosenka
            while ostatnia_piosenka is not None:
                if ostatnia_piosenka.nastepna is None:
                    ostatnia_piosenka = ostatnia_piosenka
                    break
                else:
                    ostatnia_piosenka = ostatnia_piosenka.nastepna
            ostatnia_piosenka.nastepna = PiosenkaWPlayliscie(piosenka, None, None)
            
    def odtworz(self):
        while self.aktywna_piosenka.nastepna is not None:
            print(f'gram:{self.aktywna_piosenka.tytul}')
            self.aktywna_piosenka = self.aktywna_piosenka.nastepna
        print(f'gram:{self.aktywna_piosenka.tytul}')
    
square_hammer = Piosenka("Square hammer")
lambada = Piosenka("Lambada")
macarena = Piosenka("Macarena")
run_to_the_hills = Piosenka("Run to the hills")
running_up_that_hill = Piosenka("Running up that hill")

playlista_hani = Playlista("Playlista Hani")

playlista_hani.dodaj(square_hammer)
playlista_hani.dodaj(lambada)
playlista_hani.dodaj(macarena)

playlista_hani.odtworz()
            
        
        
```
