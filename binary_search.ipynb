```python
import math

liczby = [4,6,8,13,17,20,21,33,45]
szukana_liczba = 45
srodkowa_liczba = 0
uciete_indeksy = 0

while srodkowa_liczba != szukana_liczba:
    liczba_mieszkanek = len(liczby)
    indeks_srodka_niezaokraglony = liczba_mieszkanek / 2
    indeks_srodka_zaokraglony = math.floor(indeks_srodka_niezaokraglony)
    srodkowa_liczba = liczby[indeks_srodka_zaokraglony]

    if srodkowa_liczba > szukana_liczba:
        liczby = liczby[0:indeks_srodka_zaokraglony]
    elif srodkowa_liczba < szukana_liczba:
        liczby = liczby[indeks_srodka_zaokraglony:liczba_mieszkanek]
        uciete_indeksy = uciete_indeksy + indeks_srodka_zaokraglony
    else:
        print(indeks_srodka_zaokraglony+uciete_indeksy)
    
```
