```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-100,100)

y1 = 4*x+28
y2 = 45+x     # y = 0.5x + 7

plt.xlim(0,10)
plt.ylim(45,55)

plt.xticks(np.arange(0, 10, 1)) 
plt.yticks(np.arange(45, 55, 1)) 

plt.plot(x, y1, label='y = x + 2')
plt.plot(x, y2, label='y = 45 + x')

plt.title('Dwie funkcje liniowe')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
