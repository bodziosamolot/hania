```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10)

y1 = x+2
y2 = x*0.5+3     # y = 0.5x + 7

plt.xlim(0,10)
plt.ylim(0,10)

plt.xticks(np.arange(0, 10, 1)) 
plt.yticks(np.arange(0, 10, 1)) 

plt.plot(x, y1, label='y = x + 2')
plt.plot(x, y2, label='y = 0.5x + 3')

plt.title('Dwie funkcje liniowe')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True, alpha=0.3)

plt.show()
```
