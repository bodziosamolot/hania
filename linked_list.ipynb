```python
class Uczennica:
    def __init__(self,nastepna,imie,wiek):
        self.imie = imie 
        self.nastepna = nastepna
        self.wiek = wiek
    
    def przedstaw_sie(self):
        if self.nastepna is None:
            print(f'Czesc! Jestem {self.imie}. Mam {self.wiek} lat.Nie widze przed soba nikogo.')
        else:
            print(f'Czesc! Jestem {self.imie}. Mam {self.wiek} lat.Widze przed soba {self.nastepna.imie}')
        
hania = Uczennica(None, "hania", 12)
zosia = Uczennica(hania, "zosia", 10)
gosia = Uczennica(zosia, "gosia", 18)
grapefruit = Uczennica(gosia, "grapefruit",1)

kolezanka = grapefruit
najstarsza = kolezanka
while kolezanka is not None:
    # kolezanka.przedstaw_sie()
    
    if kolezanka.nastepna is None:
        break
    if kolezanka.wiek > najstarsza.wiek:
        najstarsza = kolezanka
    
    kolezanka = kolezanka.nastepna

najstarsza.przedstaw_sie()
```
