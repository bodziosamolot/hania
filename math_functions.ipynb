```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rysuj_funkcje(*funkcje, x_min=-10, x_max=10):
    x = np.linspace(x_min, x_max, 500)

    plt.figure()
    for f, nazwa in funkcje:
        y = f(x)
        plt.plot(x, y, label=nazwa)

    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Wykres funkcji")
    plt.show()
```

```python
liniowa = lambda x: 2 * x + 1
liniowa_plus_jeden = lambda x: 2 * x + 1 + 1
liniowa_przesunieta_w_prawo = lambda x: 2 * (x+1) + 1 + 1
rysuj_funkcje((liniowa, "y = 2x + 1"), (liniowa_plus_jeden, "y = 2x + 2"), (liniowa_przesunieta_w_prawo, "y = 2x + -1 + 1 +1"), x_min=-2, x_max=2)
```

```python
kwadratowa = lambda x: x ** 2
rysuj_funkcje((kwadratowa, "y = x²"), x_min=-3, x_max=3)
```

```python
x2 = lambda x: x*x
x3 = lambda x: x*x+1
przesuniete_w_prawo = lambda x: (x-1)*(x-1)+1
rysuj_funkcje((x2, "y = x²"), (x3, "y = x²+1"), (przesuniete_w_prawo, "przesuniete_w_prawo"), x_min=-1.5, x_max=1.5)
```

```python
sinus = lambda x: np.sin(x)
cosinus = lambda x: np.cos(x)
rysuj_funkcje((sinus, "y = sin(x)"), (cosinus, "y = cos(x)"), x_min=-2 * np.pi, x_max=2 * np.pi)
```
