```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```

```python
def rysuj_funkcje(f, x_min=-10, x_max=10, nazwa="f(x)"):
    x = np.linspace(x_min, x_max, 500)
    y = f(x)

    plt.figure()
    plt.plot(x, y, label=nazwa)
    plt.axhline(y=0, color='black', linewidth=0.5)
    plt.axvline(x=0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(nazwa)
    plt.show()
```

```python
# y = 2x + 1
rysuj_funkcje(lambda x: 2 * x + 1, x_min=-5, x_max=5, nazwa="y = 2x + 1")
```

```python
# y = x²
rysuj_funkcje(lambda x: x ** 2, x_min=-5, x_max=5, nazwa="y = x²")
```

```python
# y = sin(x)
rysuj_funkcje(lambda x: np.sin(x), x_min=-2 * np.pi, x_max=2 * np.pi, nazwa="y = sin(x)")
```
