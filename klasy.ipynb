```python
class Osoba:
    def __init__(self, imie, nazwisko, wiek):
        self.imie = imie 
        self.nazwisko = nazwisko
        self.wiek = wiek

    def przedstaw_sie(self):
        print(f'Czesc! Jestem {self.imie} {self.nazwisko}. Mam {self.wiek} lat.')

Hania = Osoba("Hania", "Ogieglo", 12)
Zuzia = Osoba("Zuzia", "Ballion", 12)
Malgosia = Osoba("Malgosia", "Plaza", 12)

Hania.przedstaw_sie()
Zuzia.przedstaw_sie()
Malgosia.przedstaw_sie()
```

```python
class kwadrat:
    def __init__(self,dlugosc_boku):
        self.dlugosc_boku = dlugosc_boku
    
    def oblicz_pole(self):
        pole = self.dlugosc_boku**2
        return pole
    
    def oblicz_obwod (self):
        obwod = self.dlugosc_boku * 4
        return obwod

class trojkat_rownoboczny:
    def __init__(self,wysokosc,dlugosc_boku):
        self.wysokosc = wysokosc
        self.dlugosc_boku = dlugosc_boku

    def oblicz_pole(self):
        pole = self.wysokosc * self.dlugosc_boku / 2
        return pole
    
    def oblicz_obwod(self):
        obwod = self.dlugosc_boku * 3
        return obwod
```
