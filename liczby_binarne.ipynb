```python
def na_liczbe_binarna(liczba_dziesietna):
    liczba_binarna=""
    while liczba_dziesietna / 2 >= 0.5:
        if liczba_dziesietna % 2 == 1:
            liczba_binarna = liczba_binarna + '1'
        else:
            liczba_binarna = liczba_binarna + '0'
        
        liczba_dziesietna = liczba_dziesietna // 2
    return "".join(reversed(liczba_binarna))

wynik = na_liczbe_binarna(16)
print(wynik)
```
